```python
import os
import gradio as gr
from pathlib import Path
from dotenv import load_dotenv
from openai import OpenAI

# 1. Dynamically locate and safely load the .env file
notebook_dir = Path(os.getcwd())
root_dir = notebook_dir.parent if notebook_dir.name == "Projects" else notebook_dir
env_path = root_dir / ".env"

if env_path.exists():
    load_dotenv(dotenv_path=env_path)

# 2. Initialize the Client
gemini_api_key = os.getenv("GEMINI_API_KEY")
gemini_base_url = os.getenv("GEMINI_BASE_URL", "https://generativelanguage.googleapis.com/v1beta/openai/")

if not gemini_api_key:
    raise ValueError("❌ Missing Credentials: Make sure your key is in your .env file.")

gemini = OpenAI(base_url=gemini_base_url, api_key=gemini_api_key)

# 3. Model Logic
system_prompt = """You are a Travel assistant made for the maps of India. Help the user to find all the nearby famous places 
keeping in mind that he/she is a tourist. Also suggest hotels or rooms for stay within 2km radius of 
railway station or airport."""

def travel_assistant(message, history):
    messages = [{'role': 'system', 'content': system_prompt}]
    
    # Optional: Adds memory context if you chat back and forth in the UI
    for human, assistant in history:
        messages.append({'role': 'user', 'content': human})
        messages.append({'role': 'assistant', 'content': assistant})
        
    messages.append({'role': 'user', 'content': message})
    
    try:
        response = gemini.chat.completions.create(
            model="gemini-2.5-flash", 
            messages=messages
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"❌ Error: {e}"

# 4. Launch the Beautiful Gradio Interface
demo = gr.ChatInterface(
    fn=travel_assistant,
    title="🗺️ Indian Maps Travel Assistant",
    description="Type in your current Indian city, railway station, or airport to get tourist hotspots and nearby stays within 2km!",
    examples=[
        "I just arrived at Secunderabad Railway Station.",
        "Show me tourist spots near Vijayawada Junction.",
        "Looking for budget stays within 2km of Kochi Airport."
    ]
)

if __name__ == "__main__":
    # If running in a notebook, this will display the UI inline automatically
    demo.launch(inbrowser=True)
```
