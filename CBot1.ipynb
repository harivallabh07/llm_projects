```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr 
```

```python
""" Declaring models that we use in the app. 
1. Gemini with OpenAi SDK.
2. Ollama(Llama3) with OpenAi SDK.
3. OpenRouter with OpenAi SDK."""

load_dotenv()
gemini = OpenAI(base_url=os.getenv('GEMINI_BASE_URL'), api_key=os.getenv('GEMINI_API_KEY'))
ollama = OpenAI(base_url=os.getenv('OLLAMA_BASE_URL'), api_key=os.getenv('OLLAMA_API_KEY'))
openrouter = OpenAI(base_url=os.getenv('OPENROUTER_BASE_URL'), api_key=os.getenv('OPENROUTER_API_KEY'))
```

```python
system_message="""You are a very helpful assistant."""
```

<h3>Models</h3>

```python
def get_response(client, model, message, history):
    response = client.chat.completions.create(
        model=model,
        messages=history + [
            {"role": "system", "content": system_message},
            {"role": "user", "content": message}
        ]
    )
    return response.choices[0].message.content
```

```python
def gemini_response(message, history):
    return get_response(
        gemini,
        "gemini-2.5-flash",
        message,
        history
    )

def llama_response(message, history):
    return get_response(
        ollama,
        "llama3.2",
        message,
        history
    )

def openrouter_response(message, history):
    return get_response(
        openrouter,
        "openai/gpt-3.5-turbo",
        message,
        history
    )
```

```python
def select_model(message, history, bot):
    if bot == "gemini_chat":
        return gemini_response(message, history)

    elif bot == "ollama_chat":
        return llama_response(message, history)

    elif bot == "openrouter_chat":
        return openrouter_response(message, history)

    return "Invalid model selected."
```

# Gradio Time

```python
demo = gr.ChatInterface(
    fn=select_model,
    additional_inputs=[
        gr.Radio(
            choices=["gemini_chat", "ollama_chat",'openrouter_chat'],
            label="Select Bot",
            value="gemini_chat"
        )
    ]
)
demo.launch(inbrowser=True)
```
