```python
import os
from pathlib import Path
from dotenv import load_dotenv 
from openai import OpenAI
import gradio as gr
from IPython.display import Markdown
```

```python
gemini_api_key = os.getenv("GEMINI_API_KEY")
gemini_base_url = os.getenv("GEMINI_BASE_URL")
```

```python
gemini = OpenAI(base_url=gemini_base_url, api_key=gemini_api_key)
```

```python
system_prompt="""
You are a helpful travel assistant bulit for Indian roads. Find the hotels and restaurants near the railway stations and airports.
You can also answer any questions the user may have about their trip."""
```

```python
def gemini_chat(message):
    response = gemini.chat.completions.create(
        model="gemini-3.1-flash-lite",
        messages=[{'role': 'system', 'content': system_prompt},{'role': 'user', 'content': message}]
    )
    return (response.choices[0].message.content)
```

```python
ollama_base_url = os.getenv("OLLAMA_BASE_URL")
ollama_api_key = os.getenv("OLLAMA_API_KEY")
ollama = OpenAI(base_url=ollama_base_url, api_key=ollama_api_key)
```

```python
def ollama_chat(message):
    response = ollama.chat.completions.create(
        model="llama3.2",
        messages=[{'role': 'system', 'content': system_prompt},{'role': 'user', 'content': message}]
    )
    return (response.choices[0].message.content)
```

```python
def both_chat(message,select_bot):
    if select_bot == "gemini_chat":
        return gemini_chat(message)
    elif select_bot == "ollama_chat":
        return ollama_chat(message)
    else:
        return "Invalid bot selection"
```

# Gradio Time!!

```python
input=gr.Textbox(label="Enter your message",placeholder="Ask me anything about your trip!",lines=10)
select_bot = gr.Radio(label="Select Bot",choices=["gemini_chat","ollama_chat"])
output=gr.Markdown(label="Response")
gr.Interface(fn=both_chat,inputs=[input,select_bot],outputs=output).launch(inbrowser=True)
```
